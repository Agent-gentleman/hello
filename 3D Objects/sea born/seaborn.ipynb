```python
tips_fg=sns.FacetGrid(tips_df,col='time',row='smoker')
tips_fg.map(plt.scatter, 'total_bill','tip')
```

```python
tips_df
```

```python
tips_fg=sns.FacetGrid(tips_df,col='time',hue='smoker',palette='Set1',height=4,aspect=1.3,col_order=['Dinner','Lunch'])
tips_fg.map_dataframe(plt.scatter, 'total_bill','tip',edgecolor='w')
```

```python
kws=dict(s=50,linewidth=.5,edgecolor='w')
tips_fg=sns.FacetGrid(tips_df,col='sex',hue='smoker',height=4,aspect=1.3,
                     hue_order=['Yes','No'],
                     hue_kws=dict(marker=['^','v']))
tips_fg.map_dataframe(plt.scatter,'total_bill','tip', **kws)
```

```python
att_df=sns.load_dataset('attention')
att_fg=sns.FacetGrid(att_df,col='subject',col_wrap=5,height=1.5)
att_fg.map_dataframe(plt.plot,'solutions','score',marker='.')
```
